```python
# generate short term of transaction list
import pandas as pd
all_label = pd.read_csv('all_label.csv')
all_label['Address'] = all_label['Address'].str.lower()
ponzi_address = all_label[all_label['Class'].apply(str) == '1']
normal_address = all_label[all_label['Class'].apply(str) == '0']

# delete toAddress which are not hash address (data cleaning)
txn_all = pd.read_csv('transaction_all.csv')
txn_all = txn_all.drop(txn_all[txn_all['toAddress'].str[0:2] != '0x'].index,axis=0)

txn_all['fromAddress'] = txn_all['fromAddress'].str.lower()
txn_all['toAddress'] = txn_all['toAddress'].str.lower()

merge1 = pd.merge(txn_all,ponzi_address,left_on='toAddress',right_on='Address',how='left')
merge1 = merge1.dropna(axis=0)
merge1 = merge1[merge1['direction'] == 'IN']
merge1 = merge1.drop(['Address','Id','Class'], axis=1)
merge1['Class'] = '1'
merge1.drop_duplicates()
merge2 = pd.merge(txn_all,ponzi_address,left_on='fromAddress',right_on='Address',how='left')
merge2 = merge2.dropna(axis=0)
merge2 = merge2[merge2['direction'] == 'OUT']
merge2 = merge2.drop(['Address','Id','Class'], axis=1)
merge2['Class'] = '1'
merge2.drop_duplicates()
merge3 = pd.merge(txn_all,normal_address,left_on='toAddress',right_on='Address',how='left')
merge3 = merge3.dropna(axis=0)
merge3 = merge3[merge3['direction'] == 'IN']
merge3 = merge3.drop(['Address','Id','Class'], axis=1)
merge3['Class'] = '0'
merge3.drop_duplicates()
merge4 = pd.merge(txn_all,normal_address,left_on='fromAddress',right_on='Address',how='left')
merge4 = merge4.dropna(axis=0)
merge4 = merge4[merge4['direction'] == 'OUT']
merge4 = merge4.drop(['Address','Id','Class'], axis=1)
merge4['Class'] = '0'
merge4.drop_duplicates()
merge = merge1.append(merge2).append(merge3).append(merge4).drop_duplicates()

merge.loc[merge['direction'] == 'IN','address'] = merge.loc[merge['direction'] == 'IN','toAddress']
merge.loc[merge['direction'] == 'OUT','address'] = merge.loc[merge['direction'] == 'OUT','fromAddress']

#add toAdd&fromAdd
merge['toAdd&fromAdd'] = merge['toAddress'] +'&' + merge['fromAddress']

# add creator
import pandas as pd
creator_ac = pd.read_csv("all_label_result.csv",header=None).drop_duplicates();
df = pd.DataFrame(columns=['address','creator'])
df['creator'] = creator_ac[1]
df['address'] = creator_ac[0]
df['address'] = df['address'].str.lower()
df.set_index(['address'],inplace=True)
df = df.drop_duplicates()

merge = pd.merge(merge,creator_ac,left_on='address',right_on=creator_ac[0],right_index=False).drop([0],axis=1).rename(columns={1:'creator'});


# insert column inter_direction
internal = merge[merge['txn_type'] == 'internal']
external = merge[merge['txn_type'] == 'external']
external = external.drop(['toAdd&fromAdd','age','block','txFee','type','value','Class','txn_type','toAddress','address','direction','creator'],axis=1)
external.rename(columns={'fromAddress':'fromAddress_former'},inplace=True)
merge = pd.merge(internal,external,on="txhash",how="left")

merge.insert(len(merge.columns),'inter_direction',"2") # other
merge.loc[merge['toAddress'] == merge['creator'],'inter_direction'] = "0" # owner
merge.loc[merge['toAddress'] == merge['fromAddress_former'],'inter_direction'] = "1" # investor
merge.drop(['fromAddress_former'],axis=1,inplace=True)
external = merge[merge['txn_type'] == 'external']
merge = pd.concat([merge,external],ignore_index=True)
merge['inter_direction'].fillna("3",inplace=True)

# get sample ponzi address
a = merge[merge['Class'] == '1']
ponzi_lenth = len(a['toAdd&fromAdd'].drop_duplicates())

import numpy as np
size = 10000
z = a['toAdd&fromAdd'].drop_duplicates().sample(n=size)
z = pd.DataFrame({'toAdd&fromAdd':z})
x = pd.merge(z,a,on='toAdd&fromAdd')

# get sample non-ponzi address
b = merge[merge['Class'] == '0']
b[b['direction'] == 'IN']['toAddress'].append(b[b['direction'] == 'OUT']['fromAddress']).drop_duplicates()

c = b['toAdd&fromAdd'].drop_duplicates().sample(n=size)

c = pd.DataFrame({'toAdd&fromAdd':c})
d = pd.merge(c,b,on='toAdd&fromAdd')

txn_short = x.append(d)
txn_short.to_csv('transaction1_1.csv',index=False)
```
