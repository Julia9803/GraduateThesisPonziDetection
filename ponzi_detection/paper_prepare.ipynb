```python
import pandas as pd
basepath = "res/res3/"
_apf = "APF"
_tbf = "TBF"
_all = "ALL"
rf = "RF"
xgb = "XGBoost"
lgb = "LightGBM"

apf1 = pd.read_csv(basepath + _apf + "1" + rf + ".csv")
apf2 = pd.read_csv(basepath + _apf + "1" + xgb + ".csv")
apf3 = pd.read_csv(basepath + _apf + "1" + lgb + ".csv")
apf1.index = pd.Series([_apf + "1" + rf])
apf2.index = pd.Series([_apf + "1" + xgb])
apf3.index = pd.Series([_apf + "1" + lgb])

res = apf1.append(apf2).append(apf3)

for i in range(1,12):
    df1 = pd.read_csv(basepath + _tbf + str(i) + rf + ".csv")
    df2 = pd.read_csv(basepath + _tbf + str(i) + xgb + ".csv")
    df3 = pd.read_csv(basepath + _tbf + str(i) + lgb + ".csv")
    df1.index = pd.Series([_tbf + str(i) + rf])
    df2.index = pd.Series([_tbf + str(i) + xgb])
    df3.index = pd.Series([_tbf + str(i) + lgb])
    res = res.append(df1).append(df2).append(df3)
  
for i in range(1,12):
    df1 = pd.read_csv(basepath + _all + str(i) + rf + ".csv")
    df2 = pd.read_csv(basepath + _all + str(i) + xgb + ".csv")
    df3 = pd.read_csv(basepath + _all + str(i) + lgb + ".csv")
    df1.index = pd.Series([_all + str(i) + rf])
    df2.index = pd.Series([_all + str(i) + xgb])
    df3.index = pd.Series([_all + str(i) + lgb])
    res = res.append(df1).append(df2).append(df3)
    
print(res)
res.to_csv("res.csv")
```
