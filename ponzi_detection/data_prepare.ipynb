```python
import pandas as pd
df = pd.read_csv("transaction1.csv")
df
```

```python
tmp = pd.DataFrame(df, columns=['txhash','fromAddress','toAddress','value','Class','age']).reset_index(drop=True)
tmp.to_csv('trans_frontend_show.csv',index=False)
```

```python
tmp
```

```python
a = tmp[tmp['value'].str.contains('0 Ether')]
list1 = list(a.value)
list2 = list(tmp.value)
res = list(set(list1)^set(list2))
result = tmp[tmp.value.isin(res)].sample(100).reset_index(drop=True)
result.loc[result['Class'] == 1,'Class'] = 'ponzi'
result.loc[result['Class'] == 0,'Class'] = 'non_ponzi'
result
```

```python
result.to_csv('trans_frontend_show2.csv')
```

```python
t = result.to_json(orient='index')
```
