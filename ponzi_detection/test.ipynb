```python
import numpy as np
import pandas as pd
# get sample ponzi address
a = merge[merge['Class'] == 1]
ponzi_lenth = len(a['toAdd&fromAdd'].drop_duplicates())

size = 10000
z = a['toAdd&fromAdd'].drop_duplicates().sample(n=size)
z = pd.DataFrame({'toAdd&fromAdd':z})
x = pd.merge(z,a,on='toAdd&fromAdd')

# get sample non-ponzi address
b = merge[merge['Class'] == 0]
b[b['direction'] == 'IN']['toAddress'].append(b[b['direction'] == 'OUT']['fromAddress']).drop_duplicates()

c = b['toAdd&fromAdd'].drop_duplicates().sample(n=size)

c = pd.DataFrame({'toAdd&fromAdd':c})
d = pd.merge(c,b,on='toAdd&fromAdd')
```
